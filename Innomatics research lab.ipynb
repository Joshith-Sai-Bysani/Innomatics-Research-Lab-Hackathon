- Which city has the highest total revenue (total_amount) from Gold members? 

```python
df_new[df_new['membership'] == 'Gold'].groupby('city')['total_amount'].sum().sort_values()
```

- Which cuisine has the highest average order value across all orders?

```python
df_new.groupby('cuisine')['total_amount'].mean()
```

- How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?

```python
count = 1
for i in df_new['user_id'].unique():
    print(count, i)
    count += 1
```

```python
(df_new.groupby('user_id')['total_amount'].sum() > 1000).sum()
```
